```python
def rgb2gray(I):
    r, g, b = I[:,:,0], I[:,:,1], I[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray
```

```python
filename_train="./dataset/training_set/"
filename_test="./dataset/testing_set/"
nbr_im=600
```

```python
#read the dataset and normalize it
def stocker_images(nbr_im, filename):
    nl,nc=taille
    mat_image=np.zeros((nl,nc,nbr_im))
    for i in range(nbr_im):
        target= filename+ str(i+1) + '.png'
        img = cv2.imread(target)
        img_resized = cv2.resize(img, dsize=(120,120), interpolation=cv2.INTER_CUBIC)
        mat_image[:,:,i]=rgb2gray(img_resized)       
    return mat_image
```

```python
# mat_image=stocker_images(nbr_im, filename)
mat_train_image=stocker_images(nbr_im,filename_train)
mat_test_image=stocker_images(nbr_im,filename_test)
# mat_test_image
```

```python
def histo(I):
    [nl,nc]=np.shape(I)
    h=np.zeros(256,dtype=int)
    I_m=np.round(I)
    for i in range(0,nl):
        for j in range(0,nc):
            val=int(I_m[i][j])
            h[val]+=1
    return h
```

```python
def stocker_histogramme(mat_image):
    nl,nc,nb_im=mat_image.shape
    mat_histo=np.zeros((nb_im,256))
    for i in range(nb_im):
        mat_histo[i,:]=histo(mat_image[:,:,i])
    return mat_histo
```

```python
mat_histo_testing=stocker_histogramme(mat_test_image)
mat_histo_training=stocker_histogramme(mat_train_image)
```

```python
def dist(x,y):
    return np.sqrt(np.sum((x-y)**2))
```

```python
def stock_dist(hist1,hist2,nbr_im):
    mat_dist=np.zeros((nbr_im,nbr_im))
    for i in range(0,nbr_im):
        for j in range(0,nbr_im):
            mat_dist[i,j]=dist(hist1[i],hist2[j])
    return mat_dist
```

```python
nbr_im=600
mat_dist=stock_dist(mat_histo_testing,mat_histo_training,nbr_im)
mat_dist
```

```python

def determiner_KNN_images(index_img_requete,nbr_img_recuperes=8):
    eval_params={}
    matrice_dest=mat_index[index_img_requete,:]
    mat_index_proches=matrice_dest[0:nbr_img_recuperes] #les indices k plus proches
    image_url=''
    k=1
    for i in mat_index_proches:
        target=filename_train + str(i) +".png"
        pic=plt.imread(target)
        plt.subplot(2,4,k)
        if i>0 and i<300:
            img_status='Non infecte'
            plt.title(img_status)
        elif i>=300 and i<600:
            img_status="Infecte"
            plt.title(img_status)
        plt.imshow(pic)
        plt.savefig('static/img/' + str(index_img_requete) + '.png')
        image_url = str(index_img_requete) + '.png'
        k=k+1    
    per=np.sum(new_mat[index_img_requete,:])
    mat_binary=new_mat[index_img_requete,0:nbr_img_recuperes]
    vp=np.sum(mat_binary) #vrais postives
    fp=nbr_img_recuperes-vp  #faux positives
    fn=per-vp #faux negatives
    vn=600-nbr_img_recuperes-fn # vrais negatives
    precis=vp*100/(vp+fp)
    reca=vp*100/(vp+fn)
    spec=vn*100/(vn+fp)
    sens=vp*100/(vp+fn)
    f1_score= 2*vp*100 / (2*vp + fp + fn)
    accuracy=(vp+vn)*100/(vp+vn+fp+fn)
    eval_params['precision']=precis
    eval_params['rappel']=reca
    eval_params['sensitivity']=sens
    eval_params['specificity']=spec
    eval_params['f1_score']=f1_score
    eval_params['accuracy']=accuracy
    eval_params['image_url']=image_url
    return eval_params
```

```python
def evaluate_model(index_image):
    eval_params = determiner_KNN_images(index_image)
    return eval_params
```

```python

app = Flask(__name__)
UPLOAD_FOLDER = 'static/uploads'
app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER
app.config['MAX_CONTENT_LENGTH'] = 16 * 1024 * 1024
ALLOWED_EXTENSIONS = set(['png', 'jpg', 'jpeg', 'gif'])
eval_params={}
def allowed_file(filename):
    return '.' in filename and filename.rsplit('.', 1)[1].lower() in ALLOWED_EXTENSIONS


@app.route('/')
def home():
    return render_template('home.html',)


@app.route('/upload', methods=['POST','GET'])
def upload():
    if request.method == 'POST':
        file = request.files['file']
        filename = secure_filename(file.filename)
        file_name = filename.split(".")
        print(file_name)
        index = int(file_name[0])
        eval_params=evaluate_model(index)
        print(eval_params)
        file.save(os.path.join(app.config['UPLOAD_FOLDER'], filename))
        return jsonify(eval_params)
if __name__ == '__main__':
    app.run()
```
